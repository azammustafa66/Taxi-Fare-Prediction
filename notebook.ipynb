### Assesing the number of payments made by a particular method
According to the data dictionary, the payment method was encoded as follows:

1 = Credit card  
2 = Cash  
3 = No charge  
4 = Dispute  
5 = Unknown  
6 = Voided trip

```python
df['payment_type'].value_counts()
```

```python
# Average payment amount made by credit card
avg_amount_paid_by_credit_card = df[df['payment_type'] == 1]['total_amount'].mean()

# Average payment amount made by cash
avg_amount_paid_by_cash = df[df['payment_type'] == 2]['total_amount'].mean()

print(
    f'Average payment amount made by credit card: {avg_amount_paid_by_credit_card}\n Average payment amount made by cash: {avg_amount_paid_by_cash}'
)
```

### Assesing data by VendorID

```python
df['VendorID'].value_counts()
```

```python
# Average total amount by each vendor
df.groupby('VendorID')['total_amount'].mean()
```

```python
# Filter data according to credit card payment
credit_card_data = df[df['payment_type'] == 1]
credit_card_data['passenger_count'].value_counts()
```

```python
# Average tip amount by passenger count
credit_card_data.groupby('passenger_count')['tip_amount'].mean()
```

After looking at the dataset, the two variables that are most likely to help build a predictive model for taxi ride fares are total_amount and trip_distance because those variables show a picture of a taxi cab ride.
